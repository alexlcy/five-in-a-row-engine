```python
import tensorflow as tf
import numpy as np
import os
```

```python
mat_files_X = ["game_data/" + file for file in os.listdir("game_data/") if file.startswith("master_mat")]
X = np.concatenate([np.load(file) for file in mat_files_X])
X.shape
```

```python
mat_files_Y = ["game_data/" + file for file in os.listdir("game_data/") if file.startswith("master_move")]
Y = np.concatenate([np.load(file) for file in mat_files_Y])
Y.shape
```

```python
X_train = X[:int(len(X)*0.8)]
X_test = X[int(len(X)*0.8):]
Y_train = Y[:int(len(X)*0.8)]
Y_test = Y[int(len(X)*0.8):]
```

```python
X_train = X_train.reshape(X_train.shape[0],8,8,1)
Y_train = Y_train.reshape(Y_train.shape[0], 64)
X_test = X_test.reshape(X_test.shape[0],8,8,1)
Y_test = Y_test.reshape(Y_test.shape[0],64)
print(X_train.shape)
```

```python
model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 150x150 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(48, (3,3), activation='relu', input_shape=(8,8,1), padding = 'same'),
    tf.keras.layers.Dropout(0.5),
    
    # The second convolution
    tf.keras.layers.Conv2D(48, (3,3), activation='relu', padding = 'same'),
    tf.keras.layers.Dropout(0.5),
    
    # The third convolution
    tf.keras.layers.Conv2D(48, (2,2), activation='relu', padding = 'same'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Flatten(),
    
    # 512 neuron hidden layer
    tf.keras.layers.Dense(1024, activation='relu'),
    
    # Last layer of model
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(64, activation='softmax')])
```

```python
model.compile(loss = 'categorical_crossentropy', 
              optimizer='rmsprop', 
              metrics=['accuracy'])

history = model.fit(X_train,Y_train, 
                    epochs=100, 
                    batch_size=256,
                    validation_data = (X_test, Y_test), 
                    verbose = 1, 
                    validation_steps=3)
```
